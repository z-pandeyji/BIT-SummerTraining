```python
def tokens_per_chat(user_token, ai_tokens, multiplier=1):
  return ( user_token + ai_tokens) * multiplier
```

```python
def monthly_usage(message_per_day, tokens_per_message, days=30):
  daily_tokens = message_per_day * tokens_per_message
  monthly_tokens = daily_tokens * days
  return daily_tokens, monthly_tokens
```

```python
def calculate_cost(tokens, price_per_1k):
  return tokens / 1000 * price_per_1k
```

```python
def context_check(tokens, context_window):
  if tokens <= context_window:
    return "Yes, fits"
  else:
    return "No, too large"
```

```python
def create_report(
    mode_name,
    message_per_day,
    user_tokens,
    ai_tokens,
    price_per_1k,
    context_window,
    multiplier=1
):
  chat_tokens = tokens_per_chat(user_tokens, ai_tokens, multiplier)
  daily_tokens, monthly_tokens = monthly_usage(message_per_day, chat_tokens)

  daily_cost =   calculate_cost(daily_tokens, price_per_1k)
  monthly_cost = calculate_cost(monthly_tokens, price_per_1k)

  print("|n" + "=" * 50)
  print(mode_name)
  print("=" * 50)

  print("Tokens per chat:", chat_tokens)
  print("Daily tokens:", daily_tokens)
  print("Monthly tokens:", monthly_tokens)
  print("Daily cost $", round(daily_cost,2))
  print("Monthly cost: $", round(monthly_cost, 2))
  print("4k context fit:", context_check(chat_tokens, context_window))

  return monthly_cost
```

```python
message_per_day = 1000
user_tokens = 50
ai_tokens = 100
price_per_1k = 0.03
context_window = 4096
```

```python
normal_cost = create_report(
    mode_name= "Normal English",
    message_per_day= message_per_day,
    user_tokens = user_tokens,
    ai_tokens = ai_tokens,
    price_per_1k = price_per_1k,
    context_window = context_window,
    multiplier= 1
)
```

```python
hindi_cost = create_report(
    mode_name= "Hindi/Hinglish Mode",
    message_per_day= message_per_day,
    user_tokens = user_tokens,
    ai_tokens = ai_tokens,
    price_per_1k = price_per_1k,
    context_window = context_window,
    multiplier= 2
)
```
