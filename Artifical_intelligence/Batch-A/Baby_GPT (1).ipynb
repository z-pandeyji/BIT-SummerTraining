```python
import random
```

```python
text = """
machine learning is a part of artificial intelligence.
aritficial intelligence is changing the world.
learning from data is the heart of machine learning .
the world is changing fast because of artificial intelligence.
data is the fuel of machine learning and artificial intelligence.
""".lower()
```

```python
word = text.split()

chain = {}

for i in range(len(word) - 1):
  current_word = word[i]
  next_word = word[i + 1]

  if current_word not in chain:
    chain[current_word] = []
  chain[current_word].append(next_word)

print("artifical ke baad:", chain["artificial"])
```

```python
print("is k baad", chain["is"])
```

```python
def generate(start_word, length = 5):
  word = start_word
  result = [word]

  for _ in range(length):
    if word not in chain:
      break
    word = random.choice(chain[word])
    result.append(word)

  return " ".join(result)
```

```python
print(generate("artificial"))
```

```python
print(generate("is"))
```
