```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([3,4])
b = np.array([1,2])

print("a=", a, "| b", b)

print(a+b)
print(a-b)
print(2 * a)
```

```python
from matplotlib.colors import ColorSequenceRegistry
plt.figure(figsize=(6,6))

for v,c, name in [(a, 'blue','a'),
 (b, 'green','b'), (a+b, 'red','a+b')]:
 plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy',
            scale=1, color=c)
 plt.text(v[0], v[1], f' {name} {tuple(v)}',
          color=c)
plt.xlim(-1,8); plt.ylim(-1,8); plt.grid(); plt.axhline(0)

plt.title("Vecotrs as arrows"); plt.show()
```

```python
print("maginitude |a| =", np.linalg.norm(a))

print("Euclidean dist (a,b) =", round(np.linalg.norm(a-b),3))

print("Manhattan Dist(a,b)", np.sum(np.abs(a-b)))

print("dot(a,b) =", np.dot(a,b))
```

```python
def cosine(x,y):
  return np.dot(x,y) / (np.linalg.norm(x) * np.linalg.norm(y))

print("cosine=", round(cosine(a,b), 4))
```

```python
p = np.array([1,1])
q = np.array([10,10])

print("cosine =", round(cosine(p,q), 3))

print("euclidean =", round(np.linalg.norm(p-q), 2))
```
