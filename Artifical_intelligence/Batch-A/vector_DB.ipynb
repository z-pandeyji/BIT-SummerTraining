```python
import chromadb
chroma = chromadb.Client()

col = chroma.create_collection('products')
```

```python
products = [
    ("Wireless headphones with noise cancellation", "electronics"),
    ("Running shoes for marathon training", "sports"),
    ("Organic green tea, 100 bags", "grocery"),
    ("Bluetooth speaker, waterproof", "electronics"),
    ("Yoga mat, non-slip", "sports")
]

col.add(
    ids = [str(i) for i in range(len(products))],
    embeddings = [embed(p[0]) for p in products],
    documents = [p[0] for p in products],
    metadatas=[{"category":p[1] for p in products}]
)

print("Stored", col.count(), "products")
```
