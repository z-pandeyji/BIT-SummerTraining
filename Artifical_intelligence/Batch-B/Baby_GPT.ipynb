```python
import random
```

```python
text = """
machine learning is a part of artificial intelligence.
artificial intelligence is changing the world.
learning from data is the heart of machine learning.
the world is changing fast because of artififcial intelligence.
data is the fuel of machine learning and artificial intelligence.
""".lower()
```

```python
words = text.split()
chains ={}

for i in range(len(words) - 1):
  chains.setdefault(words[i], []).append(words[i + 1])
```

```python
print("artificial ke baad:", chains["artificial"])
print("'is' ke baad:", chains["is"])
```

```python
print(chains)
```

```python
def generate(start_word, length= 15):
  word = start_word
  result = [word]

  for _ in range(length):
    if word not in chains:
      break
    word = random.choice(chains[word])
    result.append(word)
  return " ".join(result)
```

```python
print(generate("artificial"))
```
