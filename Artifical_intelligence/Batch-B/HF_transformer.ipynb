```python
import torch
from transformers import pipeline, set_seed
```

```python
device = 0 if torch.cuda.is_available() else -1
```

```python
print("Ready - ab asli AI models!")
print("GPU Available:", torch.cuda.is_available())
print("Device:", "GPU" if device == 0 else "CPU")
```

```python
# Sentiment AI

sentiment = pipeline(
    "sentiment-analysis",
    model= "distilbert/distilbert-base-uncased-finetuned-sst-2-english",
    device= device
)
```

```python
print(sentiment("This course is absolutely amazing!"))
print(sentiment("The service was terrible and slow."))
```

```python
print(sentiment(" Blah Blha Lorem Ipsum"))
```

```python
reviews = [
    "best Decision Ever!",
    "I want a refund.",
    "It was okay, nothing special"
]

for r in reviews:
  result = sentiment(r)[0]
  print(r, "->",
        result["label"],
        "| score",
        round(result["score"], 3))
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2",
    device=device
)
```

```python
from torch.cuda import temperature
set_seed(42)

output = generator(
    "Artificial Intelligence will",
    max_new_tokens = 25,
    do_sample = True,
    temperature = 0.8,
    top_p = 0.95,
    pad_token_id = 50256)
```

```python
print(output[0]["generated_text"])
```

```python
prompts = [
    "In the future, students will",
    "The best way to learn coding is",
    "Gorakhpur is famous for"
]

for p in prompts:
  output = generator(
      p,
      max_new_tokens=25,
      do_sample=True,
      temperature=0.8,
      top_p = 0.95,
      pad_token_id=50256

  )

  print("prompt:", p)
  print("Generated:", output[0]["generated_text"])
  print("----")
```
