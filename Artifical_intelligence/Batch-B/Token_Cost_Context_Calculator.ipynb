```python
def estimate_tokens(text):
  return max(1, round(len(text) / 4))
```

```python
print(estimate_tokens("Hello"))
```

```python
print(estimate_tokens("Artifical Intelligence is amazing"))
```

```python
print(estimate_tokens("AI" * 100))
```

```python
def estimate_cost(tokens, price_per_1k = 0.03):
  return round(tokens / 1000 * price_per_1k , 4)
```

```python
print(estimate_cost(100000))
```

```python
windows = {
    "GPT 3.5 (4k)": 4096,
    "GPT-4 (8k)": 8192,
    "GPT-4 TUrbo (128k)": 128000,
    "Claude (200K)": 200000
    }
```

```python
def analyze(text, reserve_output_token = 1000):
  t = estimate_tokens(text)
  print(f"Characters: {len(text)} | Estimated tokens: ~{t}")
  print(f"Reserve output tokens: {reserve_output_token}")
  print(f"Estimated total tokens needed: ~{t + reserve_output_token}")
  print(f"Estimated cost @$0.03/1k: ${estimate_cost(t)}")

  for name, w in windows.items():
    total_needed = t + reserve_output_token

    if total_needed <= w:
      print(f" {name}: Safe fit ")
    elif t <= w:
      print(f" {name}: input fits, but less space for answer")
    else:
      print(f" {name}: too big")
```

```python
analyze("Machine learning is transforming the world" * 2000)
```
