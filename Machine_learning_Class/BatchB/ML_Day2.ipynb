```python
X = df[['study_hours', 'attendance', 'previous_mark']]
y = df['final_marks']

print('Features', X.head())
print('Label', y.head())
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
print('Model Ready')
```

```python
naya_student = [[6, 82, 70]]

predicted = model.predict(naya_student)

# print(predicted)

print(round(predicted[0], 1))
```

```python
print("importance(weights)", model.coef_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['study_hours','attendance','previous_mark']]
y = df['final_marks']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size = 0.3, random_state = 42
)
```

```python
print(len(X_train), len(X_test))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
```

```python
preds
```

```python
y_test
```

```python
print("Galti (MAE):", round(mean_absolute_error(y_test, preds), 2))
```

```python
print('Score R2:', round(r2_score(y_test, preds),2))
```
