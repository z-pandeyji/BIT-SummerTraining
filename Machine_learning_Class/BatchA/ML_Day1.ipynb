```python
X = df[['study_hours', 'attendance', 'previous_mark']]
y = df['final_marks']

print('Features', X.head())
print('Label', y.head())
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
print('Model Ready')
```

```python
naya_student = [[6, 82, 70]]

predicted = model.predict(naya_student)

# print(predicted)

print(round(predicted[0], 1))
```

```python
print("importance(weights)", model.coef_)
```
